```python
from PIL import ImageFilter
img=Image.open('leapfrog.jpeg')

filters = [
    ImageFilter.BLUR,
    ImageFilter.CONTOUR,
    ImageFilter.DETAIL,
    ImageFilter.EDGE_ENHANCE,
    ImageFilter.EMBOSS,
    ImageFilter.FIND_EDGES,
    ImageFilter.SHARPEN,
    ImageFilter.SMOOTH
]

for f in filters:
    filtered_img = img.filter(f)
    #filtered_img.show()
    display(filtered_img)

```

```python
grayscale=img.convert("L")
croped=img.crop((10, 20, 60, 70))
upsidedown=img.transpose(Image.FLIP_LEFT_RIGHT)
display(croped)
display(upsidedown)
display(grayscale)
```
