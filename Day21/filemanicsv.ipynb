# dtypes parameter

```python

pd.read_csv('Aug Train Data.csv',usecols=['gender','target']).info(3)


pd.read_csv('Aug Train Data.csv',usecols=['gender','target'],dtype={'target':int}).info(3)
```

```python
def rename(name):
    if name == "Makati City":
        return "MC"
    else:
        return name
```

```python
#conveter

df=pd.read_csv('Zomato Data.csv',encoding='latin-1',converters={'City':rename})
df.head(3)
```
